## 2.1 Понятие объекта. Атрибуты. Методы. Встроенные типы объектов

**Объект** в Python — это базовая сущность для представления данных. </br>
К объектам в Python относятся: данные встроенных типов, встроенные и пользовательские функции, встроенные и пользовательские классы, экземпляры классов, модули и т.д. </br>
Внутренне объекты представляют собой области выделенной памяти с достаточным пространством для хранения информации об объекте. Информация об объекте делится на данные объекта (значение объекта, состояние объекта) и метаданные. </br>
*Метаданные* объекта называются атрибутами. **Атрибут** — это имя переменной или имя функции, которое связано с объектом и используется для получения информации об объекте и совершения действий над этим объектом. Все атрибуты объекта составляют *пространство имен объекта*. </br>
Доступ к атрибуту `attr` объекта `obj` производится с применением синтаксиса уточнения `obj.attr`. Доступ к атрибутам объекта является одним из самых распространенных выражений в Python

````python
list1 = [1,2]
list1.__class__
````

Атрибуты объектов могут быть именами переменных или именами функций. Если атрибут объекта является именем функции, то он называется **методом**. Для вызова функции следует указать ее имя, затем в круглых скобках через запятую перечислить аргументы вызова функции

````python
list1.append(3); print(list1)
list1.__class__ # атрибут, не являющийся методом
````

Просмотреть пространство имен объекта можно с помощью встроенной функции `dir`. Аргументом функции `dir` является экземпляр объекта или тип объекта

````python
print(dir(list1))
dir(list1) == dir(list)
````

Атрибуты, которые начинаются и заканчиваются двумя знаками подчеркивания, *чаще всего* являются **методами, реализующими операции** над объектами (*dunder method*, dunder comes from 'double underscore' или *magic methods*)

````python
print([attr for attr in dir(list1) if attr.startswith('_')])
````

**Операция** (operator) — это конструкция языка, аналогичная по записи математическим операциям. Для обозначения операции часто используют не алфавитно-цифровые, а специальные символы.

````python
a = 1
a + 2, a.__add__(2)
````

### Встроенные типы объектов

Встроенные типы объектов делятся на **числа** и **коллекции**. </br>
Типы объектов, представляющих числа: **целые числа (`int`), вещественные числа (`float`), комплексные числа (`complex`). Булевый тип (`bool`)** наследуется от типа `int`, поэтому булевый тип относится к типам числовых объектов. </br>
**Коллекции** — это наборы объектов, которые делятся на три вида: </br>
**Последовательности** — это *позиционно упорядоченные* коллекции. Встроенные типы, относящиеся к последовательностям: **строки (`str`), списки (`list`),
кортежи (`tuple`).** </br>
**Отображения** — это неупорядоченные коллекции объектов, *хранящихся по ключам*. Встроенные типы, относящиеся к отображениям: **словарь (`dict`)**. </br>
**Множества** — это неупорядоченные коллекции *не повторяющихся* объектов. Встроенные типы, относящиеся к множествам: **множество (`set`)**. </br>
Встроенная функция `type(obj)` возвращает тип объекта `obj`

````python
type(1), type('1'), type([1])
````

Встроенные типы являются частью базового языка Python и всегда доступны для использования. </br>
Встроенные типы Python задействуют уже оптимизированные алгоритмы для работы со структурами данных, которые реализованы на С, чтобы обеспечивать <u>высокое быстродействие</u>. </br>
Встроенные типы объектов не только облегчают программирование, но также являются более мощными и эффективными структурами данных, чем большинство того, что можно создать в Python самостоятельно.

## 2.11 Форматированный вывод строк

````python
# Форматированный вывод с синтаксисом f'...{variable=:...}'
number = 0.123456789
result_f = f'{number=:.2%}'
print("f-string:", result_f)
# Использование бинарной операции %
result_percent = 'number=%.2f%%' % (number * 100)
print("bin %:   ", result_percent)
# Использование строкового метода format
result_format = 'number={:.2%}'.format(number)
print("format:   ", result_format)
````

````python
# Форматированный вывод с синтаксисом f'...{expression=:...}'
a, b = 5, 3
result_f = f'{(a + b)=:.3e}'
print("f-string:", result_f)
# Использование бинарной операции %
result_percent = '(a + b)=%.3e' % (a + b)
print("bin %:   ", result_percent)
# Использование строкового метода format
result_format = '(a + b)={:.3e}'.format(a + b)
print("format:   ", result_format)
````

## 2.12 Файловый объект

### Чтение файла и запись в файл

````python
# Использование функций и методов: open, close, read, readline, readlines, write, writelines
file = open('example.txt', 'w', encoding='utf-8')
file.write('First line\n')
lines = ['Second line\n', 'Third line\n', 'Fourth line\n']
file.writelines(lines)
file.close()

file = open('example.txt', 'r', encoding='utf-8')
first_line = file.readline()
next_chars = file.read(10)
remaining_lines = file.readlines()
file.close()

with open('example.txt', 'r', encoding='utf-8') as file:
    for line in file:
        print(line, end="")

with open('example.txt', 'a', encoding='utf-8') as file:
    file.write('Fifth line (appended)\n')
    file.writelines(['Sixth line\n', 'Seventh line\n'])

with open('example.txt', 'r', encoding='utf-8') as file:
    content = file.read()
    print(content)
````

````python
# 2: Причина ошибки: В Python нельзя конкатенировать объекты разных типов. Операция + для последовательностей требует, чтобы оба операнда были одного типа. 
# Здесь список пытаются сложить со строкой.
# Вариант: Преобразование строки в список
result1 = [1] + list('1')
print(f"[1] + list('1') = {result1}")
````

````python
# 3: Синтаксическая ошибка в записи. Вместо 0*tuple1 должно быть tuple1*0. Операция умножения для последовательностей требует, 
#чтобы последовательность была слева, а число повторений — справа
tuple1 = (1, 'String', [2, 3.3, 'str'])
print(f"tuple1 * 0 = {tuple1 * 0}")
print(f"0 * tuple1 = {0 * tuple1}") 
````

## 2.Как можно усекать и округлять число с плавающей точкой?
Усечение производят функции int (N) и math, trune (N), а округление — функция round (N, digits). Можно также округлять в меньшую сторону с помощью
math, floor (N) и округлять для отображения посредством операций форматирования строк.

````python
import math

# 1. Усечение (отбрасывание дробной части)
x = 3.14159
print(int(x))           # 3 - усечение через int
print(math.trunc(x))    # 3 - усечение через trunc

# 2. Округление
print(round(x, 2))      # 3.14 - до 2 знаков
print(round(x))         # 3 - до целого

# 3. Округление вниз и вверх
print(math.floor(x))    # 3 - вниз
print(math.ceil(x))     # 4 - вверх

# 4. Форматирование для отображения
print(f"{x:.2f}")       # 3.14 - форматирование строки
````

## 3.Взгляните на следующие три оператора. Изменяют ли они значение, выводимое для А?
А = "spam”
В - А
В = "shrubbery"

Нет: для А по-прежнему выводится "spam". Когда В присваивается строка
"shrubbery", все что происходит — переменная В переустанавливается для указания на новый строковый объект. А и В изначально разделяются, т.е. указывают
на тот же самый одиночный строковый объект "spam", но в Python два имени
никогда не связываются вместе. Таким образом, установка В в другой объект не
оказывает влияния на А. Кстати, то же самое справедливо, если бы последним
оператором здесь был В == В + ' shrubbery’ — конкатенация привела бы к созданию в качестве своего результата нового объекта, который затем был бы присвоен только В. Мы не можем перезаписывать строку (либо число или кортеж)
на месте, потому что строки неизменяемы.

````python
# Пример, иллюстрирующий ответ
A = "spam"
B = A          # B указывает на тот же объект "spam"
B = "shrubbery"  # B теперь указывает на новый объект

print(A)       # "spam" - A не изменилось
print(B)       # "shrubbery"

# Другой пример с конкатенацией
A = "spam"
B = A
B = B + " shrubbery"  # создается новый объект

print(A)       # "spam" - A не изменилось
print(B)       # "spam shrubbery"
````

4. ## Можно ли применять выражение среза строки для списка?
 Да. В отличие от методов выражения универсальны и применимы ко многим
типам. В данном случае выражение среза на самом деле является операцией над
последовательностями — оно работает на объектах последовательностей любого
типа, включая строки, списки и кортежи. Единственное отличие связано с тем,
что при нарезании списка обратно получается новый список.

````python
# Примеры среза для списка (как и для строки)

# Строка
text = "Python"
print(text[0:3])      # "Pyt" - первые 3 символа

# Список
fruits = ["яблоко", "банан", "апельсин", "киви", "груша"]
print(fruits[0:3])     # ["яблоко", "банан", "апельсин"] - первые 3 элемента
print(fruits[1:4])     # ["банан", "апельсин", "киви"] - со 2 по 4
print(fruits[-2:])     # ["киви", "груша"] - последние 2
print(fruits[::-1])    # весь список в обратном порядке

# Результат среза - новый список
new_list = fruits[0:2]
print(new_list)        # ["яблоко", "банан"]
````

## 5.Назовите два способа построения словаря с двумя ключами, ’ а ’ и ’Ь’, с каждым из которых ассоциировано значение 0.
Желаемый словарь создадут литеральное выражение, такое как
{'а1: 0, ’ b ’ : 0}, или серия присваиваний вроде D = { }, D [ ’ а ’ ] = 0
и D[ ’Ь’ ] = 0. Вы также можете применять более новую и простую в наборе форму ключевого слова diet (а=0, Ь=0) или более гибкую форму последовательности “ключей/значений” diet ( [ (’а’, 0) , (1 b’, 0) ] ). Либо из-за того,
что все значения одинаковы, вы можете использовать специальную форму
diet. fromkeys (’ab’, 0). В Python З.Х и 2.7 вы также можете применять включение словаря: { к: 0 for к in ’ ab1 }, хотя здесь оно снова может оказаться
излишеством.


````python
# Пять способов создания словаря {'a': 0, 'b': 0}

# 1. Литерал словаря
d1 = {'a': 0, 'b': 0}
print(d1)  # {'a': 0, 'b': 0}

# 2. Через присваивания
d2 = {}
d2['a'] = 0
d2['b'] = 0
print(d2)  # {'a': 0, 'b': 0}

# 3. Функция dict() с именованными аргументами
d3 = dict(a=0, b=0)
print(d3)  # {'a': 0, 'b': 0}

# 4. dict() со списком кортежей
d4 = dict([('a', 0), ('b', 0)])
print(d4)  # {'a': 0, 'b': 0}

# 5. fromkeys (все ключи с одинаковым значением)
d5 = dict.fromkeys(['a', 'b'], 0)
print(d5)  # {'a': 0, 'b': 0}

# Бонус: включение словаря
d6 = {k: 0 for k in 'ab'}
print(d6)  # {'a': 0, 'b': 0}
````

типа, хотя некоторые могут быть доступны для нескольких типов (например,

````python
# Определение размера (длины) кортежа

# Создаем кортеж
my_tuple = (1, 2, 3, 4, 5)

# Функция len() возвращает количество элементов
size = len(my_tuple)
print(size)  # 5

# len() работает с любыми контейнерами
print(len("Python"))        # 6 - строка
print(len([1, 2, 3]))       # 3 - список
print(len({'a': 1, 'b': 2})) # 2 - словарь

# Почему len() - функция, а не метод?
# Потому что она универсальна и работает с разными типами:
# - строки
# - списки
# - кортежи
# - словари
# - множества
# - и другие коллекции
````
