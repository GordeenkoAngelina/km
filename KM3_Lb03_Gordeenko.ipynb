## Задание 3.3. Движение секущей прямой к заданной линии на плоскости (версия 1)

```python
import sympy as sp 
# Создаём графическое окно
fig3 = plt.figure()
ax3 = plt.axes()
plt.axis([x_min-0.1, x_max+0.1, y_min-0.1, y_max+0.1])

# Глобальные переменные
point_B = None
secant_line = None

def init():
    """создаёт начальное состояние графической области перед началом анимации"""
    global point_B, secant_line
    
    curve, = ax3.plot(x_array, y_array, 'b')
    
    A = np.array([x_array[0], y_array[0]])
    point_A, = ax3.plot(A[0], A[1], 'ro')
    
    B = np.array([x_array[-1], y_array[-1]])
    point_B, = ax3.plot(B[0], B[1], 'go')
    
    secant_p = [A + (B - A)*t for t in [-2, 2]]
    secant_p = np.array(secant_p)
    secant_line, = ax3.plot(secant_p[:,0], secant_p[:,1], 'g')
    
    plt.legend([r'$y(x) = \sin(x)^5$', 'Unmovable point A', 'Movable point B', 'Secant line'])
    
    return curve, point_A, point_B, secant_line

def at_frame3(point):
    """обновляет положение подвижной точки B и секущей прямой"""
    global point_B, secant_line
    
    # обновляем положение точки B
    point_B.set_data([point[0]], [point[1]])
    
    # неподвижная точка A
    A = np.array([x_array[0], y_array[0]])
    
    # проверяем, совпали ли точки A и B
    if np.allclose(point, A):
        # если точки совпали - строим касательную
        # вычисляем производную с помощью sympy
        t_sp = sp.Symbol('t')
        y_sp = sp.sin(t_sp)**5
        dy_dx = sp.diff(y_sp, t_sp)
        slope = float(dy_dx.subs(t_sp, x_array[0]))
        
        # строим касательную прямую в точке A
        x_tangent = np.array([A[0] - 0.5, A[0] + 0.5])
        y_tangent = A[1] + slope * (x_tangent - A[0])
        secant_line.set_data(x_tangent, y_tangent)
        secant_line.set_color('r')
    else:
        # строим секущую через точки A и B
        secant_p = [A + (point - A)*t for t in [-2, 2]]
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
        secant_line.set_color('r')
    
    return point_B, secant_line

# Создаём анимацию (точка B движется от конечной точки к начальной)
anim3 = FuncAnimation(fig3, at_frame3, frames=matrix[::-1], init_func=init, repeat=False, interval=20)

plt.show()
```

## Задание 3.4 Движение секущей прямой к заданной линии на плоскости (версия 2)

```python

a_sec2 = 0.5
tmin_sec2 = 0
tmax_sec2 = 2 * np.pi
step_sec2 = (tmax_sec2 - tmin_sec2) / 100

# Параметрические функции кардиоиды
def x_sec2(t):
    return 2 * a_sec2 * (1 - np.cos(t)) * np.cos(t)

def y_sec2(t):
    return 2 * a_sec2 * (1 - np.cos(t)) * np.sin(t)

# Создаём массивы координат
t_array_sec2 = np.arange(tmin_sec2, tmax_sec2, step_sec2)
x_array_sec2 = x_sec2(t_array_sec2)
y_array_sec2 = y_sec2(t_array_sec2)

# Создаём матрицу координат (точка A движется от начальной к конечной)
matrix_sec2 = np.transpose([x_array_sec2, y_array_sec2])

# Находим границы для осей
x_min_sec2, x_max_sec2 = np.min(x_array_sec2), np.max(x_array_sec2)
y_min_sec2, y_max_sec2 = np.min(y_array_sec2), np.max(y_array_sec2)

# Создаём графическое окно
fig4 = plt.figure()
ax4 = plt.axes()
plt.axis([x_min_sec2 - 0.2, x_max_sec2 + 0.2, y_min_sec2 - 0.2, y_max_sec2 + 0.2])

# Глобальные переменные
point_A = None
secant_line = None

def init():
    """создаёт начальное состояние графической области перед началом анимации"""
    global point_A, secant_line
    
    # графический объект для исходной функции (вся кардиоида)
    curve, = ax4.plot(x_array_sec2, y_array_sec2, 'b', linewidth=2, label='Кардиоида')
    
    # неподвижная точка B (конечная)
    B = np.array([x_array_sec2[-1], y_array_sec2[-1]])
    point_B, = ax4.plot(B[0], B[1], 'ro', markersize=8, label='Неподвижная точка B')
    
    # подвижная точка A (начальная)
    A = np.array([x_array_sec2[0], y_array_sec2[0]])
    point_A, = ax4.plot(A[0], A[1], 'go', markersize=8, label='Подвижная точка A')
    
    # секущая прямая через точки A и B
    secant_p = [A + (B - A)*t for t in [-2, 2]]
    secant_p = np.array(secant_p)
    secant_line, = ax4.plot(secant_p[:,0], secant_p[:,1], 'g', linewidth=2, label='Секущая')
    
    plt.legend()
    plt.title('Движение секущей прямой (точка A движется к B)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, alpha=0.3)
    
    return curve, point_B, point_A, secant_line

def at_frame4(point):
    """обновляет положение подвижной точки A и секущей прямой"""
    global point_A, secant_line
    
    # обновляем положение точки A
    point_A.set_data([point[0]], [point[1]])
    
    # неподвижная точка B (конечная)
    B = np.array([x_array_sec2[-1], y_array_sec2[-1]])
    
    # проверяем, совпали ли точки A и B
    if np.allclose(point, B):
        # если точки совпали - строим касательную в точке B
        # вычисляем производную с помощью sympy
        
        # Создаём символьную переменную
        t_sp = sp.Symbol('t')
        
        # Параметрические функции символьные
        x_sp = 2 * a_sec2 * (1 - sp.cos(t_sp)) * sp.cos(t_sp)
        y_sp = 2 * a_sec2 * (1 - sp.cos(t_sp)) * sp.sin(t_sp)
        
        # Вычисляем производные dx/dt и dy/dt
        dx_dt = sp.diff(x_sp, t_sp)
        dy_dt = sp.diff(y_sp, t_sp)
        
        # Производная dy/dx = (dy/dt) / (dx/dt)
        # Подставляем t = tmax (конечная точка)
        dy_dx_val = float((dy_dt / dx_dt).subs(t_sp, tmax_sec2))
        
        # Строим касательную прямую в точке B
        # Уравнение касательной: y = y_B + k * (x - x_B)
        x_tangent = np.array([B[0] - 0.5, B[0] + 0.5])
        y_tangent = B[1] + dy_dx_val * (x_tangent - B[0])
        secant_line.set_data(x_tangent, y_tangent)
        secant_line.set_color('r')
        secant_line.set_label('Касательная')
    else:
        # строим секущую через точки A и B
        secant_p = [point + (B - point)*t for t in [-2, 2]]
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
        secant_line.set_color('r')
        secant_line.set_label('Секущая')
    
    return point_A, secant_line

# Создаём анимацию (точка A движется от начальной к конечной)
anim4 = FuncAnimation(fig4, at_frame4, frames=matrix_sec2, init_func=init, repeat=False, interval=20)

plt.show()
```

## ЛБ03. Анимация. Общие понятия

### 1. Определение анимации в программировании;
Анимация — это последовательная смена кадров (состояний графических объектов) с определённой задержкой, создающая иллюзию движения. В matplotlib это реализуется через функцию FuncAnimation, которая в каждом кадре вызывает пользовательскую функцию для обновления графических объектов.
### 2.Спецификация функции FuncAnimation;

`FuncAnimation(fig, func, frames=None, init_func=None, fargs=None, save_count=None, repeat=True, interval=200, blit=False, **kwargs) `

- Назначение: создание анимации путём последовательного вызова функции обновления кадров

- Результат: объект `FuncAnimation`

### 3. Обязательные аргументы;

- fig — графическое окно

- func — функция для обновления кадра

### 4. Подготовка аргумента frames;

frames — это итерируемый объект (список, массив, матрица). Для анимации построения — матрица координат [N x 2], где N — количество кадров. Каждый элемент — точка (x, y), которая добавляется в текущем кадре.

### 5. Откуда берутся данные для кадра;

FuncAnimation последовательно перебирает элементы из frames и передаёт их в функцию func. В каждом кадре из frames извлекается очередной элемент.

### 6. Как узнать количество кадров;

Количество кадров равно длине последовательности frames: len(frames). Можно вычислить как (tmax - tmin) / step + 1.

### 7. Технологические отличия заданий 3.1, 3.2, 3.3;

## Сравнение заданий 3.1, 3.2, 3.3

| Характеристика | Задание 3.1 | Задание 3.2 | Задание 3.3 |
|---------------|-------------|-------------|-------------|
| **Название** | Анимация построения линии | Движение точки вдоль кривой | Движение секущей к касательной |
| **Статические объекты** | Нет | Вся кривая | Кривая, неподвижная точка A |
| **Динамические объекты** | Линия (строится) | Точка | Точка B, секущая прямая |
| **Данные для кадров** | Матрица всех точек `[N×2]` | Матрица положений точки `[N×2]` | Матрица положений точки B `[N×2]` |
| **Направление обхода** | От начала к концу | От начала к концу | От конца к началу (к A) |
| **Функция обновления** | Добавляет точку к линии | Устанавливает новое положение точки | Обновляет точку и пересчитывает секущую |
| **Особенность** | Накопление координат | Только текущее положение | При совпадении точек — касательная |

### 8. Визуальное отличие кадров в 3.1;
В каждом следующем кадре к линии добавляется новая точка. Реализуется через line.set_data() с накоплением координат.

### 9. Визуальное отличие кадров в 3.2;

В каждом следующем кадре точка перемещается в новое положение вдоль кривой. Реализуется через point.set_data() с координатами текущей точки.

### 10. Визуальное отличие кадров в 3.3;

В каждом следующем кадре точка B приближается к A, а секущая меняет наклон. При совпадении секущая превращается в касательную. Реализуется через обновление точки B и пересчёт прямой через две точки.
