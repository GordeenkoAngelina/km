## Этап 3. Определение функции перемещения и нахождения угла запуска тела

Положим, что начало прямоугольной декартовой системы координат соответствует уровню земли.Тогда в начальный момент движения тела $t = 0$ имеем, что 

$$s_{x}(0) = 0, s_{y}(0) = h_{start}$$.

Для построения функции перемещениия $s_{x}(t),s_{y}(t))$ воспользуемся вторым законом Ньютона
$$ma = F$$,
где $m$ -- масса тела, ускорение движения тела $a = (s"_{x}(t),s"_{y}(t))$ и приложенная сила $F = m(0, -g)$. В результате имеем два уравнения, что

$$s"_{x}(t) = 0,  s"_{y}(t) = -g$$.
На основании уравнений с учетом начальных условий можно записать аналитический вид для функций $s_{x}(t)$ и $s_{y}(t)$
$$s_{x}(t) = s_{x}(0) = v0_{x}t = 0 + v0_{x}t,
s_{y}(t) = s{y}(0) + v0_{y}t - \frac{g}{2}t^2$$,

где $v0 = (v0_{x}, v0_{y})$ обозначает неизвестную скорость движения тела в начальный момент времени $t = 0$.


Из второго закона Ньютона следует, что тело движется с постоянным ускорением свободного падения $g$, направленным вертикально вниз. Раскладывая движение на горизонтальную и вертикальную составляющие, получаем:
По оси $Ox$ ускорение отсутствует, поэтому движение равномерное:
$$v_{x} = v0_{x} = const, s_{x}(t) = s_{x}(0) + v0_{x}t$$
По оси $Oy$ ускорение постояннл и равно $-g$б движение равнопеременное:
$$ v_{y}(t)= v0_{y} - gt, s_{y}(t) = s_{y}(0) + v0_{y}t - \frac{gt^2}{2}.$$
Начальные скорости:
$$v0_{x}=v_{0}cos\alpha, v0_{y}=v_{0}sin\alpha.$$

```python
v0_x = s_end/T
v0_x
```

Из условия задачи имеем, что $s_{y}(T) = h_{end}$. Из этого соотношеения находим значение для $v0_{y}$


```python
v0_y = (h_end-h_start+g/2*T**2)/T
v0_y
```

Значение угла запуска тела $\alpha$ определяем из соотношения $tan \alpha = \frac{v0_{y}}{v0_{x}}$


```python
alpha = math.atan(v0_y/v0_x)
alpha, math.degrees(alpha)
```

Результирующая функция перемещения имеет следующий вид

```python
f's_x(t) = {v0_x} t'
```

```python
f's_y(t) = {h_start} + {v0_y:.6}t - {g:.6}/2*t\N{superscript two}'
```

## Выполнение задания 1.3
Подготовим необходимые данные для построения секущей прямой к траектории
движения тела.
Воспользуемся векторно-параметрическим уравнением прямой для заданной
точки , лежащей на прямой, и заданного направляющего вектора $a$ 
$$
p(t) = p_{start} + at
$$

Направляющий вектор: $$ a = p_{end}- p_{start},
p_{end} \neq p_{start}$$
Введем новые переменные для описания начальной точки $ p_{start} $ и конечной точки $ p_{end} $ через их координаты. Для новых переменных используем тип массив (`ndarray`) из `numpy`. Представление координат точек в виде массивов иногда эффективнее, чем представление в виде списков, для дальнейшей работы с координатами.

```python
p_start = np.array([s_x_array[0],s_y_array[0]])
p_end = np.array([s_x_array[-1],s_y_array[-1]])
p_start, p_end
```

Вычислим направляющий вектор:

```python
a = p_end-p_start
```

Для построения графика прямой достаточно задания двух точек, лежащих на
прямой:

```python
p = [p_start + a*t for t in [-1, 2]]
p = np.array(p)
print(p)
```

Построим график траектории движения тела и график секущей прямой в одной

```python
fig = plt.figure()
plt.plot(s_x_array,s_y_array,'b-',lw=5)
plt.plot(p[:,0],p[:,1],'--',color='red',lw=1)
plt.xlabel('x (m)'); plt.ylabel('y (m)')
plt.axis([-5, s_end+5, 0, h_end+11]);
```

Для построения касательной прямой к траектории движения тела в момент времени $t = 0$ воспользуемся уравнением прямой, проходящей через точку $p_{\text{start}}(x_{\text{start}}, y_{\text{start}})$ с угловым коэффициентом $k$:

$$
y = k(x - x_{\text{start}}) + y_{\text{start}}
$$
Угловой коэффициент можно вычислить по формуле:
$$
k = \frac{dy}{dx} = \frac{y'(t)}{x'(t)}
$$

при $t = 0$.

Признак перпендикулярности прямых: $$
k_{\text{n} = -\frac{1}{k}
$$

Тогда уравнение нормальной прямой примет вид:

$$
y = k_{\text{n}} (x - x_{\text{start}}) + y_{\text{start}}
$$

```python
t=sympy.symbols('t')
s_difx=sympy.diff(v0_x*t, t)
s_dify=sympy.diff(h_start+v0_y*t+g*t**2/2, t)
k=s_dify.subs(t,0)/s_difx.subs(t,0)
y_k1 = np.array([s_x_array[0],s_y_array[0]])
y_k2 = np.array([s_x_array[0]+1,k+s_y_array[0]])
y_k=[y_k1+ (y_k2-y_k1)*t for t in [-1, 2]]
y_k = np.array(y_k)
y_n1 = np.array([s_x_array[0],s_y_array[0]])
y_n2 = np.array([s_x_array[0]+1,-1/k+s_y_array[0]])
y_n=[y_n1+ (y_n2-y_n1)*t for t in [-6, 10]]
y_n = np.array(y_n)
```

```python
fig = plt.figure()
plt.plot(s_x_array,s_y_array,'b-',lw=5)
plt.plot(p[:,0],p[:,1],'--',color='red',lw=1)
plt.plot(y_k[:,0], y_k[:,1], color='orange', lw=3)
plt.plot(y_n[:,0],y_n[:,1],color='green',lw=3)
plt.xlabel('x (m)'); plt.ylabel('y (m)')
plt.axis([-5, s_end+5, 0, h_end+11]);
```
